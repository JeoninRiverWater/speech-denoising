```python
import os
import glob

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


from pprint import pprint
```

```python
EVALUATION_DIR = os.path.join('..', 'PROJECT','EVALUATION')
csv_files = [file for file in os.listdir(EVALUATION_DIR) if file.endswith('.csv')]
```

```python
print(csv_files)
```

```python
data = []
for file in csv_files:
    df = pd.read_csv(os.path.join(EVALUATION_DIR, file), index_col=0)
    df = df.rename(columns=lambda x: f"{file.split('_')[0]}-{x}")
    data.append(df)
```

```python
print([len(data[i]) for i in range(len(data))])
```

```python
df = pd.concat(data, axis=1)
df.head(100)
```

```python
cols = [c for c in df.columns if any(metric in c for metric in ['SDR'])]#['SDR', 'SIR', 'SAR'])]
print(cols)

print(df[cols].mean().round(2))

plt.figure()
df[cols].boxplot()
plt.show()
```

```python
def load_example(base_dir, oracles_dir, example_id):
    
    audios = {}
    for model in os.listdir(base_dir):
        model_dir = os.path.join(base_dir, model)
        
        if os.path.isdir(model_dir):
            audios.update({
                model.split('_')[0]: {
                    'mixture': os.path.join(oracles_dir, str(example_id).zfill(4), 'mixture.wav'),
                    'clean': os.path.join(model_dir, str(example_id).zfill(4), 'clean.wav'),
                    'noise': os.path.join(model_dir, str(example_id).zfill(4), 'noise.wav'),
                }
            })
    
    import librosa
    for model in audios:
        for key in audios[model]:
            audios[model][key], _ = librosa.load(audios[model][key], sr=16000)
            audios[model][key] /= np.max(np.abs(audios[model][key]))
            audios[model][key] -= np.mean(audios[model][key])
    
    return audios
            
example_dict = load_example(EVALUATION_DIR, os.path.join(EVALUATION_DIR, 'Oracle_mixes_16kHz_4s'), example_id=0)
pprint(example_dict)
```

```python
plt.figure(figsize=(30,10))
for i, model in enumerate(example_dict):
    print(model)
    
    plt.subplot(len(example_dict), 3, 3*i+1)
    plt.plot(example_dict[model]['mixture'])
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(len(example_dict), 3, 3*i+2)
    plt.plot(example_dict[model]['clean'])
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(len(example_dict), 3, 3*i+3)
    plt.plot(example_dict[model]['noise'])
    plt.xticks([])
    plt.yticks([])

# plt.show()
plt.savefig('')
```
